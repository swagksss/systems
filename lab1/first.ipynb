# Завдання 5

```python
import math

def is_prime(n):
    if(n > 1):
        for i in range(2, int(math.sqrt(n)) + 1):
            if (n % i == 0):
                return False
        return True
    return False

def task_6():
    s = input()
    values = [int(x) for x in s.split(' ')]
    a, b = values[:2]
    redundant_values = values[2:]
    if len(redundant_values):
        print(f"Ignored inputs: {', '.join(str(x) for x in redundant_values)}")
    return [i for i in range(a, b + 1) if is_prime(i)]
    
```
